```python
# Read the dataset
df = pd.read_csv(r"C:\Users\Mhd Naqeeb\Downloads\Laliga.csv")
```

```python
# Replace dashes with 0
df.replace('-', 0, inplace=True)
```

```python
new_columns = ['Pos', 'Team', 'Seasons', 'Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn', 
               'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion', 'Runner-up', 'Third', 
               'Fourth', 'Fifth', 'Sixth', 'T', 'Debut', 'Since/LastApp', 'BestPosition']
df.columns = new_columns
```

```python
# Convert the debut column to a numeric value
df['debut_year'] = pd.to_numeric(df['Debut'].str[:4], errors='coerce')

# Filter the teams that started playing between 1930-1980
teams_1930_1980 = df[(df['debut_year'] >= 1930) & (df['debut_year'] <= 1980)]['Team']

# Print the teams
print("Teams that started playing between 1930-1980:")
print(teams_1930_1980)

```

```python
# Convert 'Points' column to numeric values
df['Points'] = pd.to_numeric(df['Points'], errors='coerce')

# Drop rows with NaN values in the 'Points' column
df.dropna(subset=['Points'], inplace=True)

# Print the list of teams which came Top 5 in terms of points
top_5_teams = df.sort_values('Points', ascending=False).head(5)['Team'].tolist()
print(top_5_teams)
```

```python
def Goal_diff_count():
    with open('football.csv', 'r') as file:
        data = file.readlines()

    # Remove the first line (headers)
    data = data[1:]

    # Create a dictionary to store the team and their goal difference
    goal_diff = {}

    for line in data:
        # Split the line into a list of values
        values = line.strip().split(',')

        # Get the team name, goals scored, and goals conceded
        team = values[0]
        goals_for = int(values[5])
        goals_against = int(values[6])

        # Calculate the goal difference
        diff = goals_for - goals_against

        # Add the team and its goal difference to the dictionary
        goal_diff[team] = diff

    # Return the dictionary with all the teams and their goal differences
    return goal_diff
```

```python
# Convert GamesWon and GamesPlayed columns to numeric data types
df["GamesWon"] = pd.to_numeric(df["GamesWon"], errors="coerce")
df["GamesPlayed"] = pd.to_numeric(df["GamesPlayed"], errors="coerce")

# Create a new column "Winning Percent"
df["Winning Percent"] = round((df["GamesWon"] / df["GamesPlayed"]) * 100, 2).fillna(0)

# Find the top 5 teams with the highest winning percentage
top_teams = df.sort_values(by="Winning Percent", ascending=False).head(5)

# Print the top 5 teams
print("Top 5 teams with the highest Winning Percent:")
print(top_teams[["Team", "Winning Percent"]])
```

```python
from collections import defaultdict

# Create a defaultdict to store the total points for each position
position_points = defaultdict(int)

# Group teams based on their best position
position_teams = defaultdict(list)
for i, team in df.iterrows():
    position = team["BestPosition"]
    position_teams[position].append(team)

# Calculate the total points for each position
for position, teams in position_teams.items():
    total_points = sum(team["Points"] for team in teams)
    position_points[position] += total_points

# Print the total points for each position
for position, total_points in position_points.items():
    print(f"Position {position}: {total_points}")
```
